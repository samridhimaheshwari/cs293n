```python
import numpy as np
import pandas as pd
from scapy.all import *
from tqdm import tqdm
tqdm.pandas(desc="progress-bar")
from gensim.models import Doc2Vec
from sklearn import utils
from sklearn.model_selection import train_test_split
import gensim
from sklearn.linear_model import LogisticRegression
from gensim.models.doc2vec import TaggedDocument
import os
import shutil
```

```python
def get_header(pcap_file):
    raw_file=pcap_file.encode('utf-8')
    hex_file=raw_file.hex()
    return hex_file
```

```python
def pcap_2_hex(pcap_file):
    scapy_cap = rdpcap(pcap_file)
    file_string=""
    file_string+= get_header(pcap_file)
    for packet in scapy_cap:
        raw_packet=raw(packet)
        hex_packet= str(raw_packet.hex())
        for x in range(len(hex_packet)):
                char_gram=hex_packet[x:x+10]
                file_string+=char_gram
                file_string+="    "
    return file_string
```

```python
df = pd.DataFrame(columns=['File', 'Label'])  
df
```

```python
for file in os.listdir("/home/rutwikk/NetworkML/data_pcap_ids/ids_sample_10k/"):
    if file.endswith(".pcap"):
        print("Adding File")
        hex_file_content = pcap_2_hex("/home/rutwikk/NetworkML/data_pcap_ids/ids_sample_10k/"+file)
        df2 = {'File': hex_file_content, 'Label': (file.split(".")[0]).split("_")[1] + "_" + (file.split(".")[0]).split("_")[2]}
        df = df.append(df2, ignore_index = True)
print("DF Created")
```
