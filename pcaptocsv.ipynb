```python
import pandas as pd


dfs = []
for file in os.listdir("/home/rutwikk/NetworkML/pcap_csv_train/"):
    if file.endswith(".pcap.csv"):
        df = pd.read_csv("/home/rutwikk/NetworkML/pcap_csv_train/"+file)
        df_encoded = pd.get_dummies(df, drop_first=True)
        df_encoded['label']=(file.split(".")[0]).split("_")[1] + "_" + (file.split(".")[0]).split("_")[2]
        dfs.append(df_encoded)
print(len(dfs))
```
